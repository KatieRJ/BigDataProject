```python
x= [2600, 3000, 3200, 3600, 400]
```

```python
y = [550000, 565000, 610000, 680000, 725000]
```

```python
import statistics
mValue = statistics.mean(x)
mValue
```

```python
mValueY = statistics.mean(y)
mValueY
```

```python
Variance = 1 / (5-1) * sum((num - mValue)** 2 for num in x)
Variance
```

```python
difX = [num - mValue for num in x]
difY = [num - mValueY for num in y]



CoVariance = 1/ (5-1) * (sum(difX[num] * difY[num] for num in range(5)))
CoVariance
```

```python
slope = CoVariance / Variance
slope
```
